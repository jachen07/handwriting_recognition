```python
#create the data directory and the path for the words 
words_txt_path = '/projectnb/ds340/projects/Jason_Aryan/ascii/words.txt'
data_dir = '/projectnb/ds340/projects/Jason_Aryan/word'

# create a function to read the labels (skipping those with errors)
def get_labels(words_txt_path):
    words_list = []
    with open(words_txt_path, "r") as f:
        for line in f:
            if line[0] == "#":
                continue  
            if line.split(" ")[1] != "err":  
                words_list.append(line)
    return words_list

#get all valid lines
words_list = get_labels(words_txt_path)
print(f"Number of valid words: {len(words_list)}")

#shuffle the words for randomness and to avoid bias 
np.random.shuffle(words_list)
print(len(words_list))
```

```python
#split into training, validation, and testing sets 
split_ratio = int(0.9 * len(words_list))
training_data = words_list[:split_ratio]
evaluation_data = words_list[split_ratio:]

val_split_ratio = int(0.5 * len(evaluation_data))
validation_data = evaluation_data[:val_split_ratio]
testing_data = evaluation_data[val_split_ratio:]

#check that the split is correct 
assert len(words_list) == len(training_data) + len(validation_data) + len(testing_data)

print(f"Total training samples: {len(training_data)}")
print(f"Total validation samples: {len(validation_data)}")
print(f"Total test samples: {len(testing_data)}")
```

```python
#set the base path to be the data directory 
base_img_path = data_dir

def extract_image_paths_and_labels(dataset):
    img_paths = []
    valid_samples = []
    missing_files = []
    
    for sample in dataset:
        split_line = sample.strip().split(" ")
        
        #check the img_name ensuring that its clean and follows the path 
        img_name = split_line[0]
        img_name = img_name.strip()
        img_path = os.path.normpath(os.path.join(base_img_path, img_name + ".png"))
        
        #check if the files exists
        if os.path.exists(img_path):
            if os.path.getsize(img_path):
                img_paths.append(img_path)
                valid_samples.append(sample.strip())
            else:
                print(f"File exists but is empty: {img_path}")
        else:
            missing_files.append(img_path)

    if missing_files:
        with open("missing_files.log", "w") as log_file:
            log_file.write("\n".join(missing_files))
        print(f"Skipped {len(missing_files)} missing files. Check 'missing_files.log' for details.")
    
    return img_paths, valid_samples
```

```python
#extract the image paths and labels for training, validation, and testing
train_img_paths, train_labels = extract_image_paths_and_labels(training_data)
validation_img_paths, validation_labels = extract_image_paths_and_labels(validation_data)
test_img_paths, test_labels = extract_image_paths_and_labels(testing_data)

print(f"Training image paths: {len(train_img_paths)}, Training labels: {len(train_labels)}")
print(f"Validation image paths: {len(validation_img_paths)}, Validation labels: {len(validation_labels)}")
print(f"Test image paths: {len(test_img_paths)}, Test labels: {len(test_labels)}")
```

```python
#calculate the length and size of the training sets 
cleaned_train_labels = []
unique_characters = set()
max_label_length = 0

for label in train_labels:
    cleaned_label = label.split(" ")[-1].strip()
    for char in cleaned_label:
        unique_characters.add(char)

    max_label_length = max(max_label_length, len(cleaned_label))
    cleaned_train_labels.append(cleaned_label)

unique_characters = sorted(list(unique_characters))

print("Maximum length: ", max_label_length)
print("Vocab size: ", len(unique_characters))

#check some training samples
print("Sample cleaned labels: ", cleaned_train_labels[:10])
```

```python
#create function to preprocess the labels 
def preprocess_labels(labels):
    preprocessed_labels = []
    for label in labels:
        cleaned_label = label.split(" ")[-1].strip()
        preprocessed_labels.append(cleaned_label)
    return preprocessed_labels

#perform preprocessing on the validation and test labels 
preprocessed_validation_labels = preprocess_labels(validation_labels)
preprocessed_test_labels = preprocess_labels(test_labels)
print("Sample preprocessed validation labels: ", preprocessed_validation_labels[:10])
print("Sample preprocessed test labels: ", preprocessed_test_labels[:10])
```

```python
#autotune for data pipeline 
AUTOTUNE = tf.data.AUTOTUNE

#map the character to integers and vice versa 
char_to_int = StringLookup(vocabulary=list(unique_characters), mask_token=None)
int_to_char = StringLookup(
    vocabulary=char_to_int.get_vocabulary(), mask_token=None, invert=True
)
```

```python
augmentation_layer = tf.keras.Sequential([
#     RandomRotation(0.03),  # 0.05, 0.1 
    RandomZoom(height_factor=(-0.1, 0.1)), # -0.2, 0.2 
])
```

```python
# write a function to apply padding to the image 
def distortion_free_resize(image, target_size):
    width, height = target_size
    image = tf.image.resize(image, size=(height, width), preserve_aspect_ratio=True)

    #calculate the padding for height and weight 
    pad_height = height - ops.shape(image)[0]
    pad_width = width - ops.shape(image)[1]

    if pad_height % 2 != 0:
        pad_height_top = (pad_height // 2) + 1
        pad_height_bottom = pad_height // 2
    else:
        pad_height_top = pad_height_bottom = pad_height // 2
    if pad_width % 2 != 0:
        pad_width_left = (pad_width // 2) + 1
        pad_width_right = pad_width // 2
    else:
        pad_width_left = pad_width_right = pad_width // 2

    #apply the padding 
    image = tf.pad(
        image,
        paddings=[
            [pad_height_top, pad_height_bottom],
            [pad_width_left, pad_width_right],
            [0, 0],
    ],
    )

    image = ops.transpose(image, (1, 0, 2))
    image = tf.image.flip_left_right(image)
    return image
```

```python
#set the initial parameters and padding for images based on width and height 
batch_size = 64
padding_token = 99
image_width = 128
image_height = 32

#create function to preprocess the image
def preprocess_image(image_path, img_size=(image_width, image_height)):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_png(image, 1)
    image = distortion_free_resize(image, img_size)
    image = augmentation_layer(image)
    image = ops.cast(image, tf.float32) / 255.0
    return image
```

```python
#create a function to vectorize the labels 
def vectorize_label(label):
    label = char_to_int(tf.strings.unicode_split(label, input_encoding="UTF-8"))
    length = ops.shape(label)[0]
    pad_amount = max_label_length - length
    label = tf.pad(label, paddings=[[0, pad_amount]], constant_values=padding_token)
    return label
```

```python
#combine the previous two functions int one 
def process_images_labels(image_path, label):
    image = preprocess_image(image_path)
    label = vectorize_label(label)
    return {"image": image, "label": label}
```

```python
#prepare the dataset for model 
def prepare_dataset(image_paths, labels):
    dataset = tf.data.Dataset.from_tensor_slices((image_paths, labels)).map(
        process_images_labels, num_parallel_calls=AUTOTUNE
    )
    return dataset.batch(batch_size).cache().prefetch(AUTOTUNE)
```

```python
#seperate into training, validation, and testing sets 
train_ds = prepare_dataset(train_img_paths, cleaned_train_labels)
validation_ds = prepare_dataset(validation_img_paths, preprocessed_validation_labels)
test_ds = prepare_dataset(test_img_paths, preprocessed_test_labels)


#check the training set images 
for data in train_ds.take(1):
    images, labels = data["image"], data["label"]

    _, ax = plt.subplots(4, 4, figsize=(15, 8))

    for i in range(16):
        img = images[i]
        img = tf.image.flip_left_right(img)
        img = ops.transpose(img, (1, 0, 2))
        img = (img * 255.0).numpy().clip(0, 255).astype(np.uint8)
        img = img[:, :, 0]

        label = labels[i]
        indices = tf.gather(label, tf.where(tf.math.not_equal(label, padding_token)))
        label = tf.strings.reduce_join(int_to_char(indices))
        label = label.numpy().decode("utf-8")

        ax[i // 4, i % 4].imshow(img, cmap="gray")
        ax[i // 4, i % 4].set_title(label)
        ax[i // 4, i % 4].axis("off")

plt.show()
```

```python
#define the CTC layer to be use in the model 
class CTCLayer(keras.layers.Layer):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.loss_fn = tf.keras.backend.ctc_batch_cost

    def call(self, y_true, y_pred):
        batch_len = ops.cast(ops.shape(y_true)[0], dtype="int64")
        input_length = ops.cast(ops.shape(y_pred)[1], dtype="int64")
        label_length = ops.cast(ops.shape(y_true)[1], dtype="int64")

        input_length = input_length * ops.ones(shape=(batch_len, 1), dtype="int64")
        label_length = label_length * ops.ones(shape=(batch_len, 1), dtype="int64")
        loss = self.loss_fn(y_true, y_pred, input_length, label_length)
        self.add_loss(loss)

        return y_pred
```

```python
def build_deep_model():
    #input layer 
    input_img = keras.Input(shape=(image_width, image_height, 1), name="image")
    labels = keras.Input(name="label", shape=(None,))

    #1st Conv2D and MaxPooling2D Layer 
    x = keras.layers.Conv2D(
        32, (3, 3), activation="relu", kernel_initializer="he_normal", padding="same", name="Conv1"
    )(input_img)
    x = keras.layers.MaxPooling2D((2, 2), name="pool1")(x)  # Reduces height and width by half

    #2nd Conv2D and MaxPooling2D Layer
    x = keras.layers.Conv2D(
        64, (3, 3), activation="relu", kernel_initializer="he_normal", padding="same", name="Conv2"
    )(x)
    x = keras.layers.MaxPooling2D((2, 2), name="pool2")(x)  # Reduces further

    #3rd Conv2D and MaxPooling2D Layer 
    x = keras.layers.Conv2D(
        128, (3, 3), activation="relu", kernel_initializer="he_normal", padding="same", name="Conv3"
    )(x)
    x = keras.layers.MaxPooling2D((2, 1), name="pool3")(x)  # Adjusted pooling for better sequence alignment

    #downscale width and height for new dimensions 
    new_width = image_width // 4  
    new_height = image_height // 8 

    # reshape for LSTMs
    new_shape = (new_width, new_height * 128) 
    x = keras.layers.Reshape(target_shape=new_shape, name="reshape")(x)
    x = keras.layers.Dense(128, activation="relu", name="dense1")(x)
    x = keras.layers.Dropout(0.2)(x)

    # 2 LSTMS layers 
    x = keras.layers.Bidirectional(
        keras.layers.LSTM(128, return_sequences=True, dropout=0.25)
    )(x)
    x = keras.layers.Bidirectional(
        keras.layers.LSTM(64, return_sequences=True, dropout=0.25)
    )(x)
    
    # 1 GRU layer 
    # x = keras.layers.Bidirectional( keras.layers.GRU(64, return_sequences=True, dropout=0.3) )(x)

    # dense and CTC layer 
    x = keras.layers.Dense(len(char_to_int.get_vocabulary()) + 2, activation="softmax", name="dense2")(x)
    output = CTCLayer(name="ctc_loss")(labels, x)

    model = keras.models.Model(inputs=[input_img, labels], outputs=output, name="handwriting_recognizer")
    model.compile(optimizer=keras.optimizers.Adam())
    return model
```

```python
# Get the updated model
model = build_deep_model()
model.summary()

validation_images = []
validation_labels = []

for batch in validation_ds:
    validation_images.append(batch["image"])
    validation_labels.append(batch["label"])
```

```python
#calculate the edit distance 
def calculate_edit_distance(labels, predictions):
    sparse_labels = ops.cast(tf.sparse.from_dense(labels), dtype=tf.int64)

    #decode the prediction to be use for edit distance 
    input_len = np.ones(predictions.shape[0]) * predictions.shape[1]
    predictions_decoded = keras.ops.nn.ctc_decode(predictions, sequence_lengths=input_len)[0][0][:, :max_label_length]
    sparse_predictions = ops.cast(tf.sparse.from_dense(predictions_decoded), dtype=tf.int64)
    
    #average out the individual edit distances 
    edit_distances = tf.edit_distance(sparse_predictions, sparse_labels, normalize=False)
    return tf.reduce_mean(edit_distances)
```

```python
#callback to compute the edit distances 
class EditDistanceCallback(keras.callbacks.Callback):
    def __init__(self, pred_model):
        super().__init__()
        self.prediction_model = pred_model

    def on_epoch_end(self, epoch, logs=None):
        edit_distances = []

        for i in range(len(validation_images)):
            labels = validation_labels[i]
            predictions = self.prediction_model.predict(validation_images[i])
            edit_distances.append(calculate_edit_distance(labels, predictions).numpy())

        mean_edit_distance = np.mean(edit_distances) 
        print(f"Mean edit distance for epoch {epoch + 1}: {mean_edit_distance:.4f}") 
```

```python
#epoch as a parameters from 1, 2, 5, 8, 10, 20, 35, 50 
#best result is 50 epochs 
epochs = 50

model = build_deep_model()
prediction_model = keras.models.Model(
    model.get_layer(name="image").output, model.get_layer(name="dense2").output
)
edit_distance_callback = EditDistanceCallback(prediction_model)

# train the model
history = model.fit(
    train_ds,
    validation_data=validation_ds,
    epochs=epochs,
    callbacks=[edit_distance_callback],
)
```

```python
#decode the batch prediction from the utility 
def decode_batch_predictions(pred):
    input_len = np.ones(pred.shape[0]) * pred.shape[1]
    results = keras.ops.nn.ctc_decode(pred, sequence_lengths=input_len)[0][0][:, :max_label_length]
    output_text = []
    for res in results:
        res = tf.gather(res, tf.where(tf.math.not_equal(res, -1)))
        res = tf.strings.reduce_join(int_to_char(res)).numpy().decode("utf-8").replace("[UNK]", "")
        output_text.append(res)
    return output_text
```

```python
#check the prediction results 
for batch in test_ds.take(1):
    batch_images = batch["image"]
    _, ax = plt.subplots(4, 4, figsize=(15, 8))

    preds = prediction_model.predict(batch_images)
    pred_texts = decode_batch_predictions(preds)

    for i in range(16):
        img = batch_images[i]
        img = tf.image.flip_left_right(img)
        img = ops.transpose(img, (1, 0, 2))
        img = (img * 255.0).numpy().clip(0, 255).astype(np.uint8)
        img = img[:, :, 0]

        title = f"Prediction: {pred_texts[i]}"
        ax[i // 4, i % 4].imshow(img, cmap="gray")
        ax[i // 4, i % 4].set_title(title)
        ax[i // 4, i % 4].axis("off")

plt.savefig('deep_model_epoch_50')
plt.show()
```
